#Q1.2:  display basic statistics and data types for each datasets

```python
# SUMMARY STATISTICS FOR NUMERICAL COLUMNS

def detailed_summary(df, dataset_name):
    print("\n" + "="*60)
    print(f"Summary Statistics for: {dataset_name}")
    print("="*60)

    # Select only numerical columns
    numeric_df = df.select_dtypes(include=['int64', 'float64'])

    summary = pd.DataFrame({
        "Mean": numeric_df.mean(),
        "Median": numeric_df.median(),
        "Std Dev": numeric_df.std(),
        "Min": numeric_df.min(),
        "25%": numeric_df.quantile(0.25),
        "50% (Median)": numeric_df.quantile(0.50),
        "75%": numeric_df.quantile(0.75),
        "Max": numeric_df.max()
    })

    display(summary)


# Applying to all datasets
detailed_summary(state_txn, "State_Txn and Users")
detailed_summary(state_split, "State_TxnSplit")
detailed_summary(state_device, "State_DeviceData")
detailed_summary(district_txn, "District_Txn and Users")
detailed_summary(district_demo, "District Demographics")
```

```python
# DISPLAY DATA TYPES FOR EACH DATASET

def display_dtypes(df, dataset_name):
    print("\n" + "="*40)
    print(f"Data Types for: {dataset_name}")
    print("="*40)
    display(df.dtypes)


# Applying to all datasets
display_dtypes(state_txn, "State_Txn and Users")
display_dtypes(state_split, "State_TxnSplit")
display_dtypes(state_device, "State_DeviceData")
display_dtypes(district_txn, "District_Txn and Users")
display_dtypes(district_demo, "District Demographics")
```

# Q1.3: Check for missing values

```python
# IDENTIFY MISSING VALUES IN EACH DATASET

def check_missing_values(df, dataset_name):
    print("\n" + "="*40)
    print(f"Missing Values in: {dataset_name}")
    print("="*40)

    missing = df.isnull().sum()
    print(missing)


# Apply to all datasets
check_missing_values(state_txn, "State_Txn and Users")
check_missing_values(state_split, "State_TxnSplit")
check_missing_values(state_device, "State_DeviceData")
check_missing_values(district_txn, "District_Txn and Users")
check_missing_values(district_demo, "District Demographics")
```

```python
# COLUMN WITH HIGHEST PERCENTAGE OF MISSING VALUES(calculating percentage and highlighting the columns)

def highest_missing_column(df, dataset_name):
    print("\n" + "="*50)
    print(f"Highest Missing Value Column: {dataset_name}")
    print("="*50)

    # Calculate missing percentage
    missing_percent = (df.isnull().sum() / len(df)) * 100

    # Filter columns with missing values
    missing_percent = missing_percent[missing_percent > 0]

    #printing missing values percentage
    print(missing_percent)

    if len(missing_percent) > 0:
        highest_col = missing_percent.idxmax()
        highest_val = missing_percent.max()

        print(f"Column with highest missing values: {highest_col}")
        print(f"Missing Percentage: {highest_val:.2f}%")
    else:
        print("No missing values found.")


# Apply to all datasets
highest_missing_column(state_txn, "State_Txn and Users")
highest_missing_column(state_split, "State_TxnSplit")
highest_missing_column(state_device, "State_DeviceData")
highest_missing_column(district_txn, "District_Txn and Users")
highest_missing_column(district_demo, "District Demographics")
```

#Q2.5: average transaction value(ATV) for each state

```python
# Calculate total transactions and total amount per state
state_avg_txn = state_txn.groupby('State').agg({
    'Transactions':'sum',
    'Amount (INR)':'sum'
}).reset_index()

# Calculate ATV
state_avg_txn['ATV'] = state_avg_txn['Amount (INR)'] / state_avg_txn['Transactions']


# Top 5 states with highest ATV
top5_highest_atv = state_avg_txn.sort_values('ATV', ascending=False).head(5)

print("Top 5 States with Highest ATV")
display(top5_highest_atv[['State','ATV']])


# Top 5 states with lowest ATV
top5_lowest_atv = state_avg_txn.sort_values('ATV', ascending=True).head(5)

print("Top 5 States with Lowest ATV")
display(top5_lowest_atv[['State','ATV']])
```
